```python
# Importar dependencias
import pandas as pd
```

```python
# Rutas de archivos
input_file = '../data/processed/dataset_ml_ready.csv'
output_file = '../data/processed/dataset_reducido_30p.csv'
```

```python
# Parametros de procesamiento por chunks

chunk_size = 10000
sample_fraction = 0.3 # Fracción de muestreo por chunk opcional para reducir tamaño
first_chunk = True
```

```python
# Procesar el archivo en chunks

for chunk in pd.read_csv(input_file, chunksize=chunk_size):
    # Tomar muestra aleatoria del porcentaje especificado
    sample = chunk.sample(frac=sample_fraction, random_state=42)
    sample['parameters_count'] = sample['parameters'].apply(lambda x: len(eval(x)) if pd.notnull(x) else 0)
    sample = sample.drop(columns=['parameters'])
    if first_chunk:
        sample.to_csv(output_file, index=False, mode='w')
        first_chunk = False
    else:
        sample.to_csv(output_file, index=False, mode='a', header=False)

print(f"Dataset procesado y reducido guardado en: {output_file}")
```
